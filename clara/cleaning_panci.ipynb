```python
import ast
import pandas as pd
import re

# ------------------------------------------------------------------
# Your existing classify_cookware function (copied for completeness)
# ------------------------------------------------------------------
def classify_cookware(row):
    """
    Classify a product as cookware type based on breadcrumb.
    Returns one of:
        - 'Rice Cooker'
        - 'Pressure Cooker'
        - 'Deep Fryer'
        - 'Frying Pan / Wok'
        - 'Sauce Pan / Milk Pan'
        - 'Stock Pot / Soup Pot'
        - 'Steamer / Dandang'
        - 'Cookware Set'
        - 'Accessories & Utensils'
        - 'Other Cookware'
        - 'Non-Cookware'
    """
    bread = row.get('breadcrumb')
    if pd.isna(bread):
        return 'Non-Cookware'

    try:
        parts = ast.literal_eval(bread)
    except:
        return 'Non-Cookware'

    # Remove leading "Shopee" if present
    if parts and parts[0] == 'Shopee':
        parts = parts[1:]

    # If after removal the list is empty, fallback
    if not parts:
        return 'Non-Cookware'

    # Join all parts into one lowercase string for keyword searching
    full_text = ' '.join(parts).lower()

    # Quick check: if none of the cookware‑related keywords appear, it's non‑cookware
    cookware_indicators = [
        'panci', 'wajan', 'cookware', 'masak', 'dapur',
        'rice cooker', 'magic com', 'penanak nasi', 'presto', 'pressure cooker',
        'deep fryer', 'penggorengan', 'kukus', 'dandang', 'steamer',
        'saucepan', 'stockpot', 'soup pot', 'set', 'kitchen', 'alat masak'
    ]
    if not any(kw in full_text for kw in cookware_indicators):
        return 'Non-Cookware'

    # Priority‑ordered mapping (longer keywords first)
    mapping = [
        # Rice cooker
        ('rice cooker', 'Rice Cooker'),
        ('magic com', 'Rice Cooker'),
        ('penanak nasi', 'Rice Cooker'),
        ('magicom', 'Rice Cooker'),
        # Pressure cooker
        ('presto', 'Pressure Cooker'),
        ('pressure cooker', 'Pressure Cooker'),
        # Deep fryer
        ('deep fryer', 'Deep Fryer'),
        ('penggorengan kentang', 'Deep Fryer'),
        # Frying pan / wok
        ('wajan', 'Frying Pan / Wok'),
        ('wok', 'Frying Pan / Wok'),
        ('frypan', 'Frying Pan / Wok'),
        ('penggorengan', 'Frying Pan / Wok'),
        ('frying pan', 'Frying Pan / Wok'),
        ('teflon', 'Frying Pan / Wok'),
        # Sauce pan / milk pan
        ('panci susu', 'Sauce Pan / Milk Pan'),
        ('milk pot', 'Sauce Pan / Milk Pan'),
        ('saucepan', 'Sauce Pan / Milk Pan'),
        ('sauce pan', 'Sauce Pan / Milk Pan'),
        # Stock pot / soup pot
        ('stockpot', 'Stock Pot / Soup Pot'),
        ('panci sup', 'Stock Pot / Soup Pot'),
        ('soup pot', 'Stock Pot / Soup Pot'),
        ('panci sayur', 'Stock Pot / Soup Pot'),
        ('panci rebus', 'Stock Pot / Soup Pot'),
        ('stock pot', 'Stock Pot / Soup Pot'),
        # Steamer / dandang
        ('steamer', 'Steamer / Dandang'),
        ('kukus', 'Steamer / Dandang'),
        ('dandang', 'Steamer / Dandang'),
        ('langseng', 'Steamer / Dandang'),
        ('klakat', 'Steamer / Dandang'),
        # Cookware set
        ('set', 'Cookware Set'),
        ('paket', 'Cookware Set'),
        ('cookware set', 'Cookware Set'),
        # Accessories & utensils
        ('tutup', 'Accessories & Utensils'),
        ('rak', 'Accessories & Utensils'),
        ('gagang', 'Accessories & Utensils'),
        ('saringan', 'Accessories & Utensils'),
        ('alat', 'Accessories & Utensils'),
        ('aksesoris', 'Accessories & Utensils'),
        ('spatula', 'Accessories & Utensils'),
        ('pisau', 'Accessories & Utensils'),
        ('parutan', 'Accessories & Utensils'),
    ]

    for keyword, cat in mapping:
        if keyword in full_text:
            return cat

    # If we reach here but it's still cookware, use a generic category
    return 'Other Cookware'

# ------------------------------------------------------------------
# Apply classification and drop fishing rows
# ------------------------------------------------------------------
# Assume df is your DataFrame with columns 'breadcrumb' and 'title'

# Step 1: Classify
df['cookware_type'] = df.apply(classify_cookware, axis=1)

# Step 2: Create a boolean mask for rows to drop
# Drop rows that are Non-Cookware AND title contains "pancing"
drop_mask = (df['cookware_type'] == 'Non-Cookware') & (df['title'].str.contains('pancing', case=False, na=False))

# Step 3: Drop those rows
df_filtered = df[~drop_mask].copy()

# (Optional) Drop the temporary classification column if you don't need it
# df_filtered = df_filtered.drop(columns=['cookware_type'])

# Check results
print(f"Original rows: {len(df)}")
print(f"Rows dropped (fishing): {drop_mask.sum()}")
print(f"Remaining rows: {len(df_filtered)}")
print("\nCookware type distribution after filtering:")
print(df_filtered['cookware_type'].value_counts())
```

```python
import pandas as pd
import re

def categorize_jenis_garansi(val):
    """Categorise warranty type into standard groups."""
    if pd.isna(val):
        return 'No Warranty'
    s = str(val).strip().lower()

    # 1. No Warranty
    no_warranty = [
        'tidak garansi', 'tidak ada garansi', 'tidak ada', '-',
        '0', 'tdk garansi', 'tidak tersedia'
    ]
    if any(kw in s for kw in no_warranty):
        return 'No Warranty'

    # 2. Manufacturer Warranty
    manufacturer = [
        'garansi produsen', 'garansi resmi', 'resmi',
        'garansi resmi pabrik', 'garansi stein mito seluruh indonesia'
    ]
    if any(kw in s for kw in manufacturer):
        return 'Manufacturer Warranty'

    # 3. Supplier Warranty (includes distributor)
    supplier = [
        'garansi supplier', 'garansi distributor', 'garansi resmi distributor'
    ]
    if any(kw in s for kw in supplier):
        return 'Supplier Warranty'

    # 4. Seller Warranty
    seller = [
        'garansi toko', 'toko', 'seller warranty', 'store warranty',
        'shopee', 'store'
    ]
    if any(kw in s for kw in seller):
        return 'Seller Warranty'

    # 5. International Warranty
    international = ['garansi internasional']
    if any(kw in s for kw in international):
        return 'International Warranty'

    # 6. Default fallback
    return 'No Warranty'
```

```python
df['Jenis Garansi'] = df['Jenis Garansi'].apply(categorize_jenis_garansi)

# Check distribution
print(df['Jenis Garansi'].value_counts())
```

```python
import pandas as pd
import re

def categorize_cookware_material(val):
    if pd.isna(val):
        return 'Other'

    s = str(val).strip().lower()

    # ---- Base materials (priority order) ----
    # Stainless Steel
    stainless_keywords = ['stainless', 'inox', 'sus304', 'sus316', 'stenlis', 'steel', 'stainless steel']
    if any(kw in s for kw in stainless_keywords):
        return 'Stainless Steel'

    # Aluminium
    alu_keywords = ['aluminium', 'aluminum', 'alumunium', 'alu']
    if any(kw in s for kw in alu_keywords):
        return 'Aluminium'

    # Cast Iron
    cast_iron_keywords = ['cast iron', 'besi cor']
    if any(kw in s for kw in cast_iron_keywords):
        return 'Cast Iron'

    # Carbon Steel
    carbon_steel_keywords = ['baja karbon', 'carbon steel', 'baja']
    if any(kw in s for kw in carbon_steel_keywords):
        return 'Carbon Steel'

    # Ceramic / Stoneware
    ceramic_keywords = ['keramik', 'stoneware', 'tanah liat', 'batu', 'stone ware']
    if any(kw in s for kw in ceramic_keywords):
        return 'Ceramic / Stoneware'

    # Glass
    glass_keywords = ['kaca', 'glass']
    if any(kw in s for kw in glass_keywords):
        return 'Glass'

    # Enamel
    enamel_keywords = ['enamel', 'email']
    if any(kw in s for kw in enamel_keywords):
        return 'Enamel'

    # ---- Coatings / non‑stick (only if no base material was found) ----
    nonstick_keywords = [
        'teflon', 'marble', 'granit', 'granite', 'anti lengket', 'coating',
        'terlapisi', 'greblon', 'lapisan', 'maifan', 'batu medis', 'ceramic coating',
        'marble coating', 'granite coating', 'non-stick', 'nonstick'
    ]
    if any(kw in s for kw in nonstick_keywords):
        return 'Non-Stick Coated'

    # ---- Catch all ----
    return 'Other'

df['Bahan'] = df['Bahan'].apply(categorize_cookware_material)

# Check the distribution
print(df['Bahan'].value_counts())
```

```python
import pandas as pd
import re

# -------------------------------------------------------------------
# 1. Clean seller_chat_time_reply to numeric hours
# -------------------------------------------------------------------

def clean_chat_time(val):
    if pd.isna(val):
        return None
    s = str(val).strip().lower()
    if s == 'hitungan jam':
        return 1.0
    if s == 'hitungan menit':
        return 0.5
    match = re.search(r'([\d,\.]+)\s*(jam|menit|hours?|hrs?)', s)
    if match:
        num_str = match.group(1).replace(',', '.')
        try:
            num = float(num_str)
        except:
            return None
        unit = match.group(2)
        if unit.startswith('jam') or unit.startswith('hour') or unit.startswith('hr'):
            return num
        elif unit.startswith('menit'):
            return num / 60.0
        else:
            return None
    try:
        return float(s)
    except:
        return None

df['seller_chat_time_reply'] = df['seller_chat_time_reply'].apply(clean_chat_time)

# -------------------------------------------------------------------
# 2. Clean seller_joined_date to numeric years (relative to 2026-03-21)
# -------------------------------------------------------------------

REF_DATE = pd.to_datetime('2026-03-21')

def parse_absolute(s):
    s_clean = re.sub(r'\s*GMT.*$', '', s)
    s_clean = re.sub(r'\s*\(.*$', '', s_clean)
    try:
        return pd.to_datetime(s_clean, errors='coerce')
    except:
        return pd.NaT

def relative_to_years(s):
    match = re.match(r'(\d+)\s+(tahun|bulan|hari)\s+lalu', s)
    if not match:
        return None
    num = int(match.group(1))
    unit = match.group(2)
    if unit == 'tahun':
        return float(num)
    elif unit == 'bulan':
        return num / 12.0
    else:  # hari
        return num / 365.25

def convert_seller_joined_date(val):
    if pd.isna(val):
        return None
    val_str = str(val).strip()
    if 'lalu' in val_str:
        years = relative_to_years(val_str)
        if years is not None:
            return years
    dt = parse_absolute(val_str)
    if pd.notna(dt):
        return (REF_DATE - dt).days / 365.25
    return None

df['seller_joined_date'] = df['seller_joined_date'].apply(convert_seller_joined_date)

```

```python
import pandas as pd
import re

def categorize_warranty(val):
    # 1. Handle missing values
    if pd.isna(val):
        return 'No Warranty'

    s = str(val).strip().lower()

    # 2. No Warranty keywords
    no_warranty = [
        'tidak garansi', 'tidak ada', '0', 'tdk garansi',
        'tidak tersedia', 'tidak ada garansi'
    ]
    if any(kw in s for kw in no_warranty):
        return 'No Warranty'

    # 3. Platform / Seller warranties → Short Term
    if 'shopee' in s or 'penjual' in s:
        return 'Short Term'

    # 4. Parse numeric duration
    match = re.search(r'(\d+(?:\.\d+)?)\s*(hari|jam|minggu|bulan|tahun)', s)
    if match:
        num = float(match.group(1))
        unit = match.group(2)

        # Convert to days for comparison
        if unit == 'jam':
            days = num / 24.0
        elif unit == 'hari':
            days = num
        elif unit == 'minggu':
            days = num * 7
        elif unit == 'bulan':
            days = num * 30.44
        else:  # tahun
            days = num * 365.25

        # Categorize by duration
        if unit in ['hari', 'jam'] and days <= 3:
            return 'Immediate'
        elif unit == 'minggu':
            return 'Short Term'
        elif unit == 'hari' and days > 3:
            return 'Short Term'
        elif unit == 'bulan' and num <= 6:
            return 'Medium Term'
        elif unit == 'bulan' and num > 6:
            return 'Long Term'
        elif unit == 'tahun':
            return 'Long Term'

    # 5. Fallback
    return 'No Warranty'

# Apply to your DataFrame column
# Replace 'warranty' with the actual column name
df['Masa Garansi'] = df['Masa Garansi'].apply(categorize_warranty)

# Check distribution
print(df['Masa Garansi'].value_counts())
```

```python
import pandas as pd
import re
import ast

# ==================================================================
# Constants & Mapping Dictionaries
# ==================================================================

REF_DATE = pd.to_datetime('2026-03-21')
THRESHOLD_BRAND = 0.001   # 0.1% of rows

# ------------------------------------------------------------------
# Country standardisation mapping
# ------------------------------------------------------------------
country_map = {
    'indonesia': 'Indonesia', 'INDONESIA': 'Indonesia',
    'china': 'China', 'CHINA': 'China', 'cina': 'China',
    'import': 'Import', 'IMPORT': 'Import', 'impor': 'Import',
    'lainnya': 'Lainnya', 'LAINNYA': 'Lainnya', 'lainya': 'Lainnya',
    'korea': 'Korea', 'jepang': 'Jepang', 'JEPANG': 'Jepang',
    'thailand': 'Thailand', 'THAILAND': 'Thailand',
    'turki': 'Turki', 'TURKI': 'Turki',
    'amerika': 'Amerika', 'AMERIKA': 'Amerika',
    'vietnam': 'Vietnam', 'VIETNAM': 'Vietnam',
    'belanda': 'Belanda', 'jerman': 'Jerman', 'prancis': 'Prancis',
    'india': 'India', 'INDIA': 'India',
    'malaysia': 'Malaysia', 'MALAYSIA': 'Malaysia',
    'brasil': 'Brasil', 'BRAZIL': 'Brasil',
    'paris': 'Prancis', 'turkiye': 'Turki',
    'germany': 'Jerman', 'swiss': 'Swiss',
    'belgium': 'Belgia', 'mexico': 'Meksiko',
}

# ------------------------------------------------------------------
# Region mapping for Dikirim Dari
# ------------------------------------------------------------------
region_map = {
    # Jabodetabek
    'KOTA JAKARTA SELATAN': 'Jabodetabek',
    'KOTA JAKARTA BARAT': 'Jabodetabek',
    'KOTA JAKARTA UTARA': 'Jabodetabek',
    'KOTA JAKARTA TIMUR': 'Jabodetabek',
    'KOTA JAKARTA PUSAT': 'Jabodetabek',
    'KOTA BEKASI': 'Jabodetabek',
    'KOTA DEPOK': 'Jabodetabek',
    'KOTA TANGERANG': 'Jabodetabek',
    'KOTA TANGERANG SELATAN': 'Jabodetabek',
    'KAB. BEKASI': 'Jabodetabek',
    'KAB. TANGERANG': 'Jabodetabek',
    'KAB. BOGOR': 'Jabodetabek',
    # Java (non‑Jabodetabek)
    'KOTA BANDUNG': 'Java',
    'KAB. BANDUNG': 'Java',
    'KAB. BANDUNG BARAT': 'Java',
    'KOTA CIREBON': 'Java',
    'KOTA TASIKMALAYA': 'Java',
    'KAB. TASIKMALAYA': 'Java',
    'KOTA SUKABUMI': 'Java',
    'KAB. CIANJUR': 'Java',
    'KAB. GARUT': 'Java',
    'KAB. MAJALENGKA': 'Java',
    'KAB. PURBALINGGA': 'Java',
    'KAB. CILACAP': 'Java',
    'KAB. BANYUMAS': 'Java',
    'KAB. PEMALANG': 'Java',
    'KAB. PEKALONGAN': 'Java',
    'KOTA PEKALONGAN': 'Java',
    'KAB. BATANG': 'Java',
    'KAB. KENDAL': 'Java',
    'KOTA SEMARANG': 'Java',
    'KAB. DEMAK': 'Java',
    'KAB. REMBANG': 'Java',
    'KAB. BLORA': 'Java',
    'KAB. PATI': 'Java',
    'KAB. JEPARA': 'Java',
    'KAB. KUDUS': 'Java',
    'KOTA SURAKARTA (SOLO)': 'Java',
    'KAB. SUKOHARJO': 'Java',
    'KAB. KARANGANYAR': 'Java',
    'KAB. WONOGIRI': 'Java',
    'KOTA YOGYAKARTA': 'Java',
    'KAB. SLEMAN': 'Java',
    'KAB. BANTUL': 'Java',
    'KAB. GUNUNG KIDUL': 'Java',
    'KAB. KULON PROGO': 'Java',
    'KOTA MALANG': 'Java',
    'KAB. MALANG': 'Java',
    'KAB. PASURUAN': 'Java',
    'KOTA PASURUAN': 'Java',
    'KOTA PROBOLINGGO': 'Java',
    'KAB. PROBOLINGGO': 'Java',
    'KOTA SURABAYA': 'Java',
    'KAB. SIDOARJO': 'Java',
    'KAB. GRESIK': 'Java',
    'KAB. LAMONGAN': 'Java',
    'KAB. BOJONEGORO': 'Java',
    'KAB. TUBAN': 'Java',
    'KAB. NGAWI': 'Java',
    'KAB. MADIUN': 'Java',
    'KOTA MADIUN': 'Java',
    'KAB. MAGETAN': 'Java',
    'KAB. PONOROGO': 'Java',
    'KAB. PACITAN': 'Java',
    'KAB. TRENGGALEK': 'Java',
    'KOTA BLITAR': 'Java',
    'KAB. BLITAR': 'Java',
    'KAB. KEDIRI': 'Java',
    'KOTA KEDIRI': 'Java',
    'KAB. NGANJUK': 'Java',
    'KAB. JOMBANG': 'Java',
    'KAB. MOJOKERTO': 'Java',
    'KOTA MOJOKERTO': 'Java',
    'KAB. BANGKALAN': 'Java',
    'KAB. PAMEKASAN': 'Java',
    'KAB. SAMPANG': 'Java',
    'KAB. SUMENEP': 'Java',
    'KAB. SERANG': 'Java',
    'KAB. PANDEGLANG': 'Java',
    'KAB. LEBAK': 'Java',
    'KOTA CILEGON': 'Java',
    'KAB. BANTEN': 'Java',
    'KOTA SERANG': 'Java',
    'KAB. KEBUMEN': 'Java',
    'KAB. GROBOGAN': 'Java',
    'KAB. TEMANGGUNG': 'Java',
    'KAB. CIAMIS': 'Java',
    'KOTA TEGAL': 'Java',
    'KAB. BANJARNEGARA': 'Java',
    'KOTA MAGELANG': 'Java',
    'KAB. SRAGEN': 'Java',
    'KAB. BONDOWOSO': 'Java',
    'KAB. LUMAJANG': 'Java',
    # Bali & Nusa Tenggara
    'KAB. BULELENG': 'Bali & Nusa Tenggara',
    'KAB. BADUNG': 'Bali & Nusa Tenggara',
    'KOTA DENPASAR': 'Bali & Nusa Tenggara',
    # Sulawesi
    'KAB. BANTAENG': 'Sulawesi',
    'KOTA MAKASSAR': 'Sulawesi',
    'KOTA MANADO': 'Sulawesi',
    'KOTA KENDARI': 'Sulawesi',
    # Kalimantan
    'KOTA PONTIANAK': 'Kalimantan',
    'KOTA SAMARINDA': 'Kalimantan',
    'KOTA BALIKPAPAN': 'Kalimantan',
    'KOTA BANJARMASIN': 'Kalimantan',
    'KOTA BANJARBARU': 'Kalimantan',
    'KAB. BANJAR': 'Kalimantan',
    'KAB. HULU SUNGAI TENGAH': 'Kalimantan',
    'KAB. HULU SUNGAI UTARA': 'Kalimantan',
    'KAB. PASER': 'Kalimantan',
    'KAB. KUTAI KARTANEGARA': 'Kalimantan',
    # Sumatra
    'KAB. TANAH DATAR': 'Sumatra',
    'KOTA PADANG': 'Sumatra',
    'KOTA PEKANBARU': 'Sumatra',
    'KOTA PALEMBANG': 'Sumatra',
    'KOTA BANDAR LAMPUNG': 'Sumatra',
    'KOTA MEDAN': 'Sumatra',
    'KOTA BATAM': 'Sumatra',
    'KOTA TANJUNG PINANG': 'Sumatra',
    'KOTA LANGSA': 'Sumatra',
    'KOTA LHOKSEUMAWE': 'Sumatra',
    'KOTA BANDA ACEH': 'Sumatra',
    'KAB. ACEH BARAT': 'Sumatra',
    'KAB. BIREUEN': 'Sumatra',
    'KAB. DELI SERDANG': 'Sumatra',
    'KAB. ASAHAN': 'Sumatra',
    'KOTA PEMATANG SIANTAR': 'Sumatra',
    'KAB. KEPULAUAN MERANTI': 'Sumatra',
    'KAB. OGAN KOMERING ULU': 'Sumatra',
    'KAB. LAMPUNG TIMUR': 'Sumatra',
    'KAB. LAMPUNG TENGAH': 'Sumatra',
    'KOTA METRO': 'Sumatra',
    'KAB. BANGKA': 'Sumatra',
    'KAB. KARIMUN': 'Sumatra',
    'KOTA JAMBI': 'Sumatra',
    'KOTA SOLOK': 'Sumatra',
    'KOTA TANJUNG BALAI': 'Sumatra',
    # Overseas
    'Luar Negeri': 'Overseas',
    None: 'Unknown'
}

# ==================================================================
# Cleaning Functions
# ==================================================================

def clean_chat_time(val):
    """Convert seller_chat_time_reply to numeric hours."""
    if pd.isna(val):
        return None
    s = str(val).strip().lower()
    if s == 'hitungan jam':
        return 1.0
    if s == 'hitungan menit':
        return 0.5
    match = re.search(r'([\d,\.]+)\s*(jam|menit|hours?|hrs?)', s)
    if match:
        num_str = match.group(1).replace(',', '.')
        try:
            num = float(num_str)
        except:
            return None
        unit = match.group(2)
        if unit.startswith('jam') or unit.startswith('hour') or unit.startswith('hr'):
            return num
        elif unit.startswith('menit'):
            return num / 60.0
        else:
            return None
    try:
        return float(s)
    except:
        return None

def clean_seller_joined_date(val):
    """Convert seller_joined_date to years since reference date."""
    if pd.isna(val):
        return None
    val_str = str(val).strip()
    # relative format
    if 'lalu' in val_str:
        match = re.match(r'(\d+)\s+(tahun|bulan|hari)\s+lalu', val_str)
        if match:
            num = int(match.group(1))
            unit = match.group(2)
            if unit == 'tahun':
                return float(num)
            elif unit == 'bulan':
                return num / 12.0
            else:  # hari
                return num / 365.25
    # absolute format
    s_clean = re.sub(r'\s*GMT.*$', '', val_str)
    s_clean = re.sub(r'\s*\(.*$', '', s_clean)
    try:
        dt = pd.to_datetime(s_clean, errors='coerce')
        if pd.notna(dt):
            return (REF_DATE - dt).days / 365.25
    except:
        pass
    return None

def classify_cookware(row):
    """Classify a product as cookware type based on breadcrumb."""
    bread = row.get('breadcrumb')
    if pd.isna(bread):
        return 'Non-Cookware'
    try:
        parts = ast.literal_eval(bread)
    except:
        return 'Non-Cookware'
    if parts and parts[0] == 'Shopee':
        parts = parts[1:]
    if not parts:
        return 'Non-Cookware'
    full_text = ' '.join(parts).lower()
    cookware_indicators = [
        'panci', 'wajan', 'cookware', 'masak', 'dapur',
        'rice cooker', 'magic com', 'penanak nasi', 'presto', 'pressure cooker',
        'deep fryer', 'penggorengan', 'kukus', 'dandang', 'steamer',
        'saucepan', 'stockpot', 'soup pot', 'set', 'kitchen', 'alat masak'
    ]
    if not any(kw in full_text for kw in cookware_indicators):
        return 'Non-Cookware'
    mapping = [
        ('rice cooker', 'Rice Cooker'), ('magic com', 'Rice Cooker'),
        ('penanak nasi', 'Rice Cooker'), ('magicom', 'Rice Cooker'),
        ('presto', 'Pressure Cooker'), ('pressure cooker', 'Pressure Cooker'),
        ('deep fryer', 'Deep Fryer'), ('penggorengan kentang', 'Deep Fryer'),
        ('wajan', 'Frying Pan / Wok'), ('wok', 'Frying Pan / Wok'),
        ('frypan', 'Frying Pan / Wok'), ('penggorengan', 'Frying Pan / Wok'),
        ('frying pan', 'Frying Pan / Wok'), ('teflon', 'Frying Pan / Wok'),
        ('panci susu', 'Sauce Pan / Milk Pan'), ('milk pot', 'Sauce Pan / Milk Pan'),
        ('saucepan', 'Sauce Pan / Milk Pan'), ('sauce pan', 'Sauce Pan / Milk Pan'),
        ('stockpot', 'Stock Pot / Soup Pot'), ('panci sup', 'Stock Pot / Soup Pot'),
        ('soup pot', 'Stock Pot / Soup Pot'), ('panci sayur', 'Stock Pot / Soup Pot'),
        ('panci rebus', 'Stock Pot / Soup Pot'), ('stock pot', 'Stock Pot / Soup Pot'),
        ('steamer', 'Steamer / Dandang'), ('kukus', 'Steamer / Dandang'),
        ('dandang', 'Steamer / Dandang'), ('langseng', 'Steamer / Dandang'),
        ('klakat', 'Steamer / Dandang'),
        ('set', 'Cookware Set'), ('paket', 'Cookware Set'), ('cookware set', 'Cookware Set'),
        ('tutup', 'Accessories & Utensils'), ('rak', 'Accessories & Utensils'),
        ('gagang', 'Accessories & Utensils'), ('saringan', 'Accessories & Utensils'),
        ('alat', 'Accessories & Utensils'), ('aksesoris', 'Accessories & Utensils'),
        ('spatula', 'Accessories & Utensils'), ('pisau', 'Accessories & Utensils'),
        ('parutan', 'Accessories & Utensils'),
    ]
    for kw, cat in mapping:
        if kw in full_text:
            return cat
    return 'Other Cookware'

def categorize_jenis_garansi(val):
    """Categorise warranty type into standard groups."""
    if pd.isna(val):
        return 'No Warranty'
    s = str(val).strip().lower()
    no_warranty = ['tidak garansi', 'tidak ada garansi', 'tidak ada', '-',
                   '0', 'tdk garansi', 'tidak tersedia']
    if any(kw in s for kw in no_warranty):
        return 'No Warranty'
    manufacturer = ['garansi produsen', 'garansi resmi', 'resmi',
                    'garansi resmi pabrik', 'garansi stein mito seluruh indonesia']
    if any(kw in s for kw in manufacturer):
        return 'Manufacturer Warranty'
    supplier = ['garansi supplier', 'garansi distributor', 'garansi resmi distributor']
    if any(kw in s for kw in supplier):
        return 'Supplier Warranty'
    seller = ['garansi toko', 'toko', 'seller warranty', 'store warranty', 'shopee', 'store']
    if any(kw in s for kw in seller):
        return 'Seller Warranty'
    if 'garansi internasional' in s:
        return 'International Warranty'
    return 'No Warranty'

def categorize_warranty(val):
    """Categorise warranty duration (Masa Garansi)."""
    if pd.isna(val):
        return 'No Warranty'
    s = str(val).strip().lower()
    no_warranty = ['tidak garansi', 'tidak ada', '0', 'tdk garansi',
                   'tidak tersedia', 'tidak ada garansi']
    if any(kw in s for kw in no_warranty):
        return 'No Warranty'
    if 'shopee' in s or 'penjual' in s:
        return 'Short Term'
    match = re.search(r'(\d+(?:\.\d+)?)\s*(hari|jam|minggu|bulan|tahun)', s)
    if match:
        num = float(match.group(1))
        unit = match.group(2)
        if unit == 'jam':
            days = num / 24.0
        elif unit == 'hari':
            days = num
        elif unit == 'minggu':
            days = num * 7
        elif unit == 'bulan':
            days = num * 30.44
        else:  # tahun
            days = num * 365.25
        if unit in ['hari', 'jam'] and days <= 3:
            return 'Immediate'
        elif unit == 'minggu':
            return 'Short Term'
        elif unit == 'hari' and days > 3:
            return 'Short Term'
        elif unit == 'bulan' and num <= 6:
            return 'Medium Term'
        elif unit == 'bulan' and num > 6:
            return 'Long Term'
        elif unit == 'tahun':
            return 'Long Term'
    return 'No Warranty'

def categorize_cookware_material(val):
    """Categorise Bahan (material) into groups."""
    if pd.isna(val):
        return 'Other'
    s = str(val).strip().lower()
    stainless = ['stainless', 'inox', 'sus304', 'sus316', 'stenlis', 'steel']
    if any(kw in s for kw in stainless):
        return 'Stainless Steel'
    alu = ['aluminium', 'aluminum', 'alumunium', 'alu']
    if any(kw in s for kw in alu):
        return 'Aluminium'
    cast_iron = ['cast iron', 'besi cor']
    if any(kw in s for kw in cast_iron):
        return 'Cast Iron'
    carbon = ['baja karbon', 'carbon steel', 'baja']
    if any(kw in s for kw in carbon):
        return 'Carbon Steel'
    ceramic = ['keramik', 'stoneware', 'tanah liat', 'batu', 'stone ware']
    if any(kw in s for kw in ceramic):
        return 'Ceramic / Stoneware'
    glass = ['kaca', 'glass']
    if any(kw in s for kw in glass):
        return 'Glass'
    enamel = ['enamel', 'email']
    if any(kw in s for kw in enamel):
        return 'Enamel'
    nonstick = ['teflon', 'marble', 'granit', 'granite', 'anti lengket', 'coating',
                'terlapisi', 'greblon', 'lapisan', 'maifan', 'batu medis',
                'ceramic coating', 'marble coating', 'granite coating',
                'non-stick', 'nonstick']
    if any(kw in s for kw in nonstick):
        return 'Non-Stick Coated'
    return 'Other'

# ==================================================================
# Main Cleaning Pipeline
# ==================================================================
def clean_dataframe(df):
    """Apply all cleaning steps to the DataFrame."""
    df = df.copy()

    # 1. Seller chat time and join date
    if 'seller_chat_time_reply' in df.columns:
        df['seller_chat_time_reply'] = df['seller_chat_time_reply'].apply(clean_chat_time)
    if 'seller_joined_date' in df.columns:
        df['seller_joined_date'] = df['seller_joined_date'].apply(clean_seller_joined_date)

    # 2. Classify cookware type and remove fishing products
    if 'breadcrumb' in df.columns:
        df['cookware_type'] = df.apply(classify_cookware, axis=1)
        drop_mask = (df['cookware_type'] == 'Non-Cookware') & \
                    (df['title'].str.contains('pancing', case=False, na=False))
        df = df[~drop_mask].copy()
        print(f"Dropped {drop_mask.sum()} fishing rows. Remaining: {len(df)}")
        # Drop one-hot encoded breadcrumb columns if they exist
        breadcrumb_dummy_cols = [col for col in df.columns if col.startswith('breadcrumb_')]
        df = df.drop(columns=breadcrumb_dummy_cols, errors='ignore')


    # 4. Negara Asal – standardise and fill missing
    if 'Negara Asal' in df.columns:
        df['Negara Asal'] = df['Negara Asal'].fillna('Unknown')
        df['Negara Asal'] = df['Negara Asal'].str.lower().replace(country_map).str.title()

    # 5. Jenis Garansi – standardise
    if 'Jenis Garansi' in df.columns:
        df['Jenis Garansi'] = df['Jenis Garansi'].apply(categorize_jenis_garansi)

    # 6. Masa Garansi – categorise duration
    if 'Masa Garansi' in df.columns:
        df['Masa Garansi'] = df['Masa Garansi'].apply(categorize_warranty)

    # 7. Bahan (material) – standardise
    if 'Bahan' in df.columns:
        df['Bahan'] = df['Bahan'].apply(categorize_cookware_material)

    # 8. Dikirim Dari – region mapping
    if 'Dikirim Dari' in df.columns:
        df['Dikirim Dari'] = df['Dikirim Dari'].map(region_map).fillna('Unknown')


    # 9. Diameter – simple fill missing with 'Unknown' (no numeric extraction)
    if 'Diameter' in df.columns:
        df['Diameter'] = df['Diameter'].fillna('Unknown')

    # 10. Stok – fill missing with 0
    if 'Stok' in df.columns:
        df['Stok'] = df['Stok'].fillna(0)

    # 11. brand – fill missing and group rare brands
    if 'brand' in df.columns:
        df['brand'] = df['brand'].fillna('Unknown')
        brand_counts = df['brand'].value_counts()
        threshold = THRESHOLD_BRAND * len(df)
        df['brand'] = df['brand'].apply(lambda x: x if brand_counts[x] >= threshold else 'Other')

    # 12. Drop unwanted columns
    cols_to_drop = [
        'category_url', 'vouchers','Merek', 'is_available', 'seller_id',
        'product_ratings', 'product_variation', 'domain', 'category_id','Jenis Panci','flash_sale','Stok','review'
    ]
    df = df.drop(columns=[c for c in cols_to_drop if c in df.columns], errors='ignore')
    rename_dict = {
        'reviews': 'review_count',          # number of reviews (original)
        'Negara Asal': 'country_of_origin',
        'Masa Garansi': 'warranty_duration',
        'Jenis Garansi': 'warranty_type',
        'Bahan': 'material',
        'Dikirim Dari': 'shipped_from',
    }

    df = df.rename(columns=rename_dict)
    return df


# ==================================================================
# Execution
# ==================================================================
if __name__ == "__main__":
    # Replace with your actual file path
    input_file = "/Users/claranatalies/Documents/year 4/FYP/COMP4501-FYP/cleaned_dataset_final/Panci_updated_cleaned.csv"
    output_file = 'panci_final_cleaned.csv'

    df_raw = pd.read_csv(input_file, low_memory=False)
    df_cleaned = clean_dataframe(df_raw)
    df_cleaned.to_csv(output_file, index=False)

    print("Cleaning complete. Output saved to:", output_file)
    print("Shape after cleaning:", df_cleaned.shape)
    print(df_cleaned.head())
```
