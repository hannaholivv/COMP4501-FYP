```python
import pandas as pd
import re

# ==================================================================
# 1. Tinggi Hak (Heel Height)
# ==================================================================
def categorize_heel_height(val):
    if pd.isna(val):
        return 'Other'
    s = str(val).strip().lower()
    # direct mappings
    if s in ['flat', 'datar / flat', 'tanpa hak', 'flat rendah']:
        return 'Flat'
    if s in ['rendah', 'rendah 2 cm', 'rendah 1,5 cm', 'rendah 2,8 cm']:
        return 'Low'
    if s in ['sedang', 'sedang (2,5 cm)', 'sedang 3 cm', 'sedang 2 cm', 'sedang 2.5cm', 'sedang 3 cm', 'sedang 4 cm']:
        return 'Medium'
    if s in ['tinggi', 'tinggi 4 cm', 'tinggi 5 cm', 'tinggi tebal sol 7 cm', 'tinggi tebal sol ±6 cm']:
        return 'High'
    if 'wedges' in s:
        return 'Wedges'
    # extract first number
    match = re.search(r'(\d+(?:\.\d+)?)', s)
    if match:
        num = float(match.group(1))
        if num <= 1:
            return 'Flat'
        elif num <= 3:
            return 'Low'
        elif num <= 6:
            return 'Medium'
        else:
            return 'High'
    return 'Other'

df['Tinggi Hak'] = df['Tinggi Hak'].apply(categorize_heel_height)

# ==================================================================
# 2. Bahan (Material)
# ==================================================================
def categorize_material(val):
    if pd.isna(val):
        return 'Other'
    s = str(val).strip().lower()
    # priority: leather > synthetic > textile > rubber/eva > plastic > jelly
    if any(kw in s for kw in ['kulit', 'leather', 'suede']):
        return 'Leather'
    if any(kw in s for kw in ['sintetis', 'pu', 'polyurethane', 'faux', 'synthetic']):
        return 'Synthetic'
    if any(kw in s for kw in ['kanvas', 'rajut', 'nilon', 'mesh', 'katun', 'polyester', 'textile']):
        return 'Textile'
    if any(kw in s for kw in ['karet', 'eva', 'phylon', 'rubber']):
        return 'Rubber/EVA/Phylon'
    if any(kw in s for kw in ['pvc', 'plastik', 'mika']):
        return 'Plastic'
    if 'jelly' in s:
        return 'Jelly'
    return 'Other'

df['Bahan'] = df['Bahan'].apply(categorize_material)

# ==================================================================
# 3. Tipe Ujung Sepatu (Toe Type)
# ==================================================================
toe_map = {
    'Ujung Terbuka': 'Open Toe',
    'Ujung Sandal Terbuka': 'Open Toe',
    'Ujung Agak Terbuka': 'Open Toe',
    'Ujung Bulat': 'Round Toe',
    'Ujung Lancip': 'Pointed Toe',
    'Ujung Persegi': 'Square Toe',
    'Ujung Tertutup': 'Closed Toe',
    'Lainnya': 'Other',
    'ujung terbuka': 'Open Toe'
}
df['Tipe Ujung Sepatu'] = df['Tipe Ujung Sepatu'].map(toe_map).fillna('Other')

# ==================================================================
# 4. Wide Fit
# ==================================================================
df['Wide Fit'] = df['Wide Fit'].map({'Ya': 'Yes', 'Tidak': 'No'}).fillna('No')

# ==================================================================
# 5. Negara Asal (Country)
# ==================================================================
country_map = {
    'Indonesia': 'Indonesia',
    'Korea': 'Korea',
    'China': 'China',
    'Thailand': 'Thailand',
    'Bangladesh': 'Bangladesh',
    'Vietnam': 'Vietnam',
    'Jepang': 'Japan',
    'India': 'India',
    'Swiss': 'Switzerland',
    'Impor': 'Import',
    'Import': 'Import',
    'Sandal Import': 'Import',
    'Sandal import': 'Import',
    'Lokal': 'Local',
    'Lainnya': 'Other'
}
df['Negara Asal'] = df['Negara Asal'].map(country_map).fillna('Other')

# ==================================================================
# 6. Model Sepatu (Model)
# ==================================================================
def categorize_model(val):
    if pd.isna(val):
        return 'Other'
    s = str(val).strip().lower()
    if 'jepit' in s or 'japit' in s or 'flip flop' in s:
        return 'Flip Flops'
    if 'slide' in s:
        return 'Slides'
    if 'mules' in s:
        return 'Mules'
    if 'wedges' in s:
        return 'Wedges'
    if 'gladiator' in s:
        return 'Gladiator'
    if 'slip-on' in s or 'slip on' in s:
        return 'Slip-On'
    if 'clogs' in s:
        return 'Clogs'
    if 'heels' in s:
        return 'Sandal Heels'
    if 'sandal' in s or 'selop' in s or 'bustong' in s:
        return 'Sandal Casual'
    return 'Other'

df['Model Sepatu'] = df['Model Sepatu'].apply(categorize_model)

# ==================================================================
# 7. Tipe Pengikat (Fastener)
# ==================================================================
def categorize_fastener(val):
    if pd.isna(val):
        return 'Other'
    s = str(val).strip().lower()
    if any(kw in s for kw in ['tanpa pengikat', 'slip-on', 'slip on', 'kokop', 'selop']):
        return 'Slip-On'
    if any(kw in s for kw in ['tali', 'strap', 'slingback']):
        return 'Strap'
    if any(kw in s for kw in ['gesper', 'gasper', 'buckle']):
        return 'Buckle'
    if any(kw in s for kw in ['velcro', 'hook & loop']):
        return 'Velcro'
    if any(kw in s for kw in ['resleting', 'zipper']):
        return 'Zipper'
    if any(kw in s for kw in ['jepit', 'japit', 'magnet']):
        return 'Snap / Clip'
    if any(kw in s for kw in ['karet', 'elastis']):
        return 'Elastic'
    return 'Other'

df['Tipe Pengikat'] = df['Tipe Pengikat'].apply(categorize_fastener)

# ==================================================================
# 8. Acara (Occasion)
# ==================================================================
occasion_map = {
    'Kasual': 'Casual',
    'Daily': 'Casual',
    'Santai': 'Casual',
    'OOTD': 'Casual',
    'Formal': 'Formal / Party',
    'Pesta': 'Formal / Party',
    'KONDANGAN': 'Formal / Party',
    'Pesta / Kondangan': 'Formal / Party',
    'Casual & Formal': 'Both',
    'Casual Formal': 'Both',
    'Semua Acara': 'Both',
    'Olahraga': 'Sport',
    'Bustong': 'Other',
    'Lainnya': 'Other'
}
df['Acara'] = df['Acara'].map(occasion_map).fillna('Other')

# ==================================================================
# Optional: check value counts after cleaning
# ==================================================================
for col in ['Tinggi Hak', 'Bahan', 'Tipe Ujung Sepatu', 'Wide Fit',
            'Negara Asal', 'Model Sepatu', 'Tipe Pengikat', 'Acara']:
    if col in df.columns:
        print(f"\n{col} after cleaning:")
        print(df[col].value_counts())
```

```python
import pandas as pd
import re

# ==================================================================
# Constants
# ==================================================================
BRAND_THRESHOLD = 0.001   # 0.1% of rows

# ------------------------------------------------------------------
# Region mapping for Dikirim Dari
# ------------------------------------------------------------------
region_map = {
    # Jabodetabek
    'KOTA JAKARTA SELATAN': 'Jabodetabek',
    'KOTA JAKARTA BARAT': 'Jabodetabek',
    'KOTA JAKARTA UTARA': 'Jabodetabek',
    'KOTA JAKARTA TIMUR': 'Jabodetabek',
    'KOTA JAKARTA PUSAT': 'Jabodetabek',
    'KOTA BEKASI': 'Jabodetabek',
    'KOTA DEPOK': 'Jabodetabek',
    'KOTA TANGERANG': 'Jabodetabek',
    'KOTA TANGERANG SELATAN': 'Jabodetabek',
    'KAB. BEKASI': 'Jabodetabek',
    'KAB. TANGERANG': 'Jabodetabek',
    'KAB. BOGOR': 'Jabodetabek',
    # Java (non‑Jabodetabek)
    'KOTA BANDUNG': 'Java',
    'KAB. BANDUNG': 'Java',
    'KAB. BANDUNG BARAT': 'Java',
    'KOTA CIREBON': 'Java',
    'KOTA TASIKMALAYA': 'Java',
    'KAB. TASIKMALAYA': 'Java',
    'KOTA SUKABUMI': 'Java',
    'KAB. CIANJUR': 'Java',
    'KAB. GARUT': 'Java',
    'KAB. MAJALENGKA': 'Java',
    'KAB. PURBALINGGA': 'Java',
    'KAB. CILACAP': 'Java',
    'KAB. BANYUMAS': 'Java',
    'KAB. PEMALANG': 'Java',
    'KAB. PEKALONGAN': 'Java',
    'KOTA PEKALONGAN': 'Java',
    'KAB. BATANG': 'Java',
    'KAB. KENDAL': 'Java',
    'KOTA SEMARANG': 'Java',
    'KAB. DEMAK': 'Java',
    'KAB. REMBANG': 'Java',
    'KAB. BLORA': 'Java',
    'KAB. PATI': 'Java',
    'KAB. JEPARA': 'Java',
    'KAB. KUDUS': 'Java',
    'KOTA SURAKARTA (SOLO)': 'Java',
    'KAB. SUKOHARJO': 'Java',
    'KAB. KARANGANYAR': 'Java',
    'KAB. WONOGIRI': 'Java',
    'KOTA YOGYAKARTA': 'Java',
    'KAB. SLEMAN': 'Java',
    'KAB. BANTUL': 'Java',
    'KAB. GUNUNG KIDUL': 'Java',
    'KAB. KULON PROGO': 'Java',
    'KOTA MALANG': 'Java',
    'KAB. MALANG': 'Java',
    'KAB. PASURUAN': 'Java',
    'KOTA PASURUAN': 'Java',
    'KOTA PROBOLINGGO': 'Java',
    'KAB. PROBOLINGGO': 'Java',
    'KOTA SURABAYA': 'Java',
    'KAB. SIDOARJO': 'Java',
    'KAB. GRESIK': 'Java',
    'KAB. LAMONGAN': 'Java',
    'KAB. BOJONEGORO': 'Java',
    'KAB. TUBAN': 'Java',
    'KAB. NGAWI': 'Java',
    'KAB. MADIUN': 'Java',
    'KOTA MADIUN': 'Java',
    'KAB. MAGETAN': 'Java',
    'KAB. PONOROGO': 'Java',
    'KAB. PACITAN': 'Java',
    'KAB. TRENGGALEK': 'Java',
    'KOTA BLITAR': 'Java',
    'KAB. BLITAR': 'Java',
    'KAB. KEDIRI': 'Java',
    'KOTA KEDIRI': 'Java',
    'KAB. NGANJUK': 'Java',
    'KAB. JOMBANG': 'Java',
    'KAB. MOJOKERTO': 'Java',
    'KOTA MOJOKERTO': 'Java',
    'KAB. BANGKALAN': 'Java',
    'KAB. PAMEKASAN': 'Java',
    'KAB. SAMPANG': 'Java',
    'KAB. SUMENEP': 'Java',
    'KAB. SERANG': 'Java',
    'KAB. PANDEGLANG': 'Java',
    'KAB. LEBAK': 'Java',
    'KOTA CILEGON': 'Java',
    'KAB. BANTEN': 'Java',
    'KOTA SERANG': 'Java',
    'KAB. KEBUMEN': 'Java',
    'KAB. GROBOGAN': 'Java',
    'KAB. TEMANGGUNG': 'Java',
    'KAB. CIAMIS': 'Java',
    'KOTA TEGAL': 'Java',
    'KAB. BANJARNEGARA': 'Java',
    'KOTA MAGELANG': 'Java',
    'KAB. SRAGEN': 'Java',
    'KAB. BONDOWOSO': 'Java',
    'KAB. LUMAJANG': 'Java',
    'KAB. SUBANG': 'Java',
    'KAB. PURWAKARTA': 'Java',
    'KOTA BATU': 'Java',
    # Bali & Nusa Tenggara
    'KAB. BULELENG': 'Bali & Nusa Tenggara',
    'KAB. BADUNG': 'Bali & Nusa Tenggara',
    'KOTA DENPASAR': 'Bali & Nusa Tenggara',
    'KAB. GIANYAR': 'Bali & Nusa Tenggara',
    'KOTA MATARAM': 'Bali & Nusa Tenggara',
    # Sulawesi
    'KAB. BANTAENG': 'Sulawesi',
    'KOTA MAKASSAR': 'Sulawesi',
    'KOTA MANADO': 'Sulawesi',
    'KOTA KENDARI': 'Sulawesi',
    'KOTA PALOPO': 'Sulawesi',
    # Kalimantan
    'KOTA PONTIANAK': 'Kalimantan',
    'KOTA SAMARINDA': 'Kalimantan',
    'KOTA BALIKPAPAN': 'Kalimantan',
    'KOTA BANJARMASIN': 'Kalimantan',
    'KOTA BANJARBARU': 'Kalimantan',
    'KAB. BANJAR': 'Kalimantan',
    'KAB. HULU SUNGAI TENGAH': 'Kalimantan',
    'KAB. HULU SUNGAI UTARA': 'Kalimantan',
    'KAB. PASER': 'Kalimantan',
    'KAB. KUTAI KARTANEGARA': 'Kalimantan',
    # Sumatra
    'KAB. TANAH DATAR': 'Sumatra',
    'KOTA PADANG': 'Sumatra',
    'KOTA PEKANBARU': 'Sumatra',
    'KOTA PALEMBANG': 'Sumatra',
    'KOTA BANDAR LAMPUNG': 'Sumatra',
    'KOTA MEDAN': 'Sumatra',
    'KOTA BATAM': 'Sumatra',
    'KOTA TANJUNG PINANG': 'Sumatra',
    'KOTA LANGSA': 'Sumatra',
    'KOTA LHOKSEUMAWE': 'Sumatra',
    'KOTA BANDA ACEH': 'Sumatra',
    'KAB. ACEH BARAT': 'Sumatra',
    'KAB. BIREUEN': 'Sumatra',
    'KAB. DELI SERDANG': 'Sumatra',
    'KAB. ASAHAN': 'Sumatra',
    'KOTA PEMATANG SIANTAR': 'Sumatra',
    'KAB. KEPULAUAN MERANTI': 'Sumatra',
    'KAB. OGAN KOMERING ULU': 'Sumatra',
    'KAB. LAMPUNG TIMUR': 'Sumatra',
    'KAB. LAMPUNG TENGAH': 'Sumatra',
    'KOTA METRO': 'Sumatra',
    'KAB. BANGKA': 'Sumatra',
    'KAB. KARIMUN': 'Sumatra',
    'KOTA JAMBI': 'Sumatra',
    'KOTA SOLOK': 'Sumatra',
    'KOTA TANJUNG BALAI': 'Sumatra',
    'KAB. LANGKAT': 'Sumatra',
    'KAB. OGAN ILIR': 'Sumatra',
    'KAB. LAMPUNG SELATAN': 'Sumatra',
    'KOTA PARIAMAN': 'Sumatra',
    # Overseas
    'Luar Negeri': 'Overseas',
    None: 'Unknown'
}

# ------------------------------------------------------------------
# Country standardisation mapping
# ------------------------------------------------------------------
country_map = {
    'Indonesia': 'Indonesia',
    'Korea': 'Korea',
    'China': 'China',
    'Thailand': 'Thailand',
    'Bangladesh': 'Bangladesh',
    'Vietnam': 'Vietnam',
    'Jepang': 'Japan',
    'India': 'India',
    'Swiss': 'Switzerland',
    'Impor': 'Import',
    'Import': 'Import',
    'Sandal Import': 'Import',
    'Sandal import': 'Import',
    'Lokal': 'Local',
    'Lainnya': 'Other'
}

# ==================================================================
# Cleaning Functions
# ==================================================================

def categorize_heel_height(val):
    if pd.isna(val):
        return 'Other'
    s = str(val).strip().lower()
    # direct mappings
    if s in ['flat', 'datar / flat', 'tanpa hak', 'flat rendah']:
        return 'Flat'
    if s in ['rendah', 'rendah 2 cm', 'rendah 1,5 cm', 'rendah 2,8 cm']:
        return 'Low'
    if s in ['sedang', 'sedang (2,5 cm)', 'sedang 3 cm', 'sedang 2 cm', 'sedang 2.5cm', 'sedang 3 cm', 'sedang 4 cm']:
        return 'Medium'
    if s in ['tinggi', 'tinggi 4 cm', 'tinggi 5 cm', 'tinggi tebal sol 7 cm', 'tinggi tebal sol ±6 cm']:
        return 'High'
    if 'wedges' in s:
        return 'Wedges'
    # extract first number
    match = re.search(r'(\d+(?:\.\d+)?)', s)
    if match:
        num = float(match.group(1))
        if num <= 1:
            return 'Flat'
        elif num <= 3:
            return 'Low'
        elif num <= 6:
            return 'Medium'
        else:
            return 'High'
    return 'Other'

def categorize_material(val):
    if pd.isna(val):
        return 'Other'
    s = str(val).strip().lower()
    if any(kw in s for kw in ['kulit', 'leather', 'suede']):
        return 'Leather'
    if any(kw in s for kw in ['sintetis', 'pu', 'polyurethane', 'faux', 'synthetic']):
        return 'Synthetic'
    if any(kw in s for kw in ['kanvas', 'rajut', 'nilon', 'mesh', 'katun', 'polyester', 'textile']):
        return 'Textile'
    if any(kw in s for kw in ['karet', 'eva', 'phylon', 'rubber']):
        return 'Rubber/EVA/Phylon'
    if any(kw in s for kw in ['pvc', 'plastik', 'mika']):
        return 'Plastic'
    if 'jelly' in s:
        return 'Jelly'
    return 'Other'

toe_map = {
    'Ujung Terbuka': 'Open Toe',
    'Ujung Sandal Terbuka': 'Open Toe',
    'Ujung Agak Terbuka': 'Open Toe',
    'Ujung Bulat': 'Round Toe',
    'Ujung Lancip': 'Pointed Toe',
    'Ujung Persegi': 'Square Toe',
    'Ujung Tertutup': 'Closed Toe',
    'Lainnya': 'Other',
    'ujung terbuka': 'Open Toe'
}

def categorize_model(val):
    if pd.isna(val):
        return 'Other'
    s = str(val).strip().lower()
    if 'jepit' in s or 'japit' in s or 'flip flop' in s:
        return 'Flip Flops'
    if 'slide' in s:
        return 'Slides'
    if 'mules' in s:
        return 'Mules'
    if 'wedges' in s:
        return 'Wedges'
    if 'gladiator' in s:
        return 'Gladiator'
    if 'slip-on' in s or 'slip on' in s:
        return 'Slip-On'
    if 'clogs' in s:
        return 'Clogs'
    if 'heels' in s:
        return 'Sandal Heels'
    if 'sandal' in s or 'selop' in s or 'bustong' in s:
        return 'Sandal Casual'
    return 'Other'

def categorize_fastener(val):
    if pd.isna(val):
        return 'Other'
    s = str(val).strip().lower()
    if any(kw in s for kw in ['tanpa pengikat', 'slip-on', 'slip on', 'kokop', 'selop']):
        return 'Slip-On'
    if any(kw in s for kw in ['tali', 'strap', 'slingback']):
        return 'Strap'
    if any(kw in s for kw in ['gesper', 'gasper', 'buckle']):
        return 'Buckle'
    if any(kw in s for kw in ['velcro', 'hook & loop']):
        return 'Velcro'
    if any(kw in s for kw in ['resleting', 'zipper']):
        return 'Zipper'
    if any(kw in s for kw in ['jepit', 'japit', 'magnet']):
        return 'Snap / Clip'
    if any(kw in s for kw in ['karet', 'elastis']):
        return 'Elastic'
    return 'Other'

occasion_map = {
    'Kasual': 'Casual',
    'Daily': 'Casual',
    'Santai': 'Casual',
    'OOTD': 'Casual',
    'Formal': 'Formal / Party',
    'Pesta': 'Formal / Party',
    'KONDANGAN': 'Formal / Party',
    'Pesta / Kondangan': 'Formal / Party',
    'Casual & Formal': 'Both',
    'Casual Formal': 'Both',
    'Semua Acara': 'Both',
    'Olahraga': 'Sport',
    'Bustong': 'Other',
    'Lainnya': 'Other'
}

# ==================================================================
# Main Cleaning Pipeline
# ==================================================================
def clean_sandals(df):
    """Apply all cleaning steps to the sandals DataFrame."""
    df = df.copy()

    # 1. Drop unwanted columns
    cols_to_drop = [
        'category_url', 'vouchers', 'is_available', 'seller_id',
        'product_ratings', 'product_variation', 'domain', 'category_id', 'flash_sale','Stok'
    ]
    df = df.drop(columns=[c for c in cols_to_drop if c in df.columns], errors='ignore')

    # 2. Drop one-hot encoded breadcrumb columns
    breadcrumb_dummy_cols = [col for col in df.columns if col.startswith('breadcrumb_')]
    df = df.drop(columns=breadcrumb_dummy_cols, errors='ignore')

    # 3. Brand grouping
    if 'brand' in df.columns:
        df['brand'] = df['brand'].fillna('Unknown')
        brand_counts = df['brand'].value_counts()
        threshold = BRAND_THRESHOLD * len(df)
        df['brand'] = df['brand'].apply(lambda x: x if brand_counts[x] >= threshold else 'Other')

    # 4. Stock fill
    if 'stock' in df.columns:
        df['stock'] = df['stock'].fillna(0)

    # 5. Location (Dikirim Dari) – region mapping
    if 'Dikirim Dari' in df.columns:
        df['Dikirim Dari'] = df['Dikirim Dari'].map(region_map).fillna('Unknown')

    # 6. Shoe-specific columns
    # Tinggi Hak
    if 'Tinggi Hak' in df.columns:
        df['Tinggi Hak'] = df['Tinggi Hak'].apply(categorize_heel_height)

    # Bahan
    if 'Bahan' in df.columns:
        df['Bahan'] = df['Bahan'].apply(categorize_material)

    # Tipe Ujung Sepatu
    if 'Tipe Ujung Sepatu' in df.columns:
        df['Tipe Ujung Sepatu'] = df['Tipe Ujung Sepatu'].map(toe_map).fillna('Other')

    # Wide Fit
    if 'Wide Fit' in df.columns:
        df['Wide Fit'] = df['Wide Fit'].map({'Ya': 'Yes', 'Tidak': 'No'}).fillna('No')

    # Negara Asal
    if 'Negara Asal' in df.columns:
        df['Negara Asal'] = df['Negara Asal'].map(country_map).fillna('Other')

    # Model Sepatu
    if 'Model Sepatu' in df.columns:
        df['Model Sepatu'] = df['Model Sepatu'].apply(categorize_model)

    # Tipe Pengikat
    if 'Tipe Pengikat' in df.columns:
        df['Tipe Pengikat'] = df['Tipe Pengikat'].apply(categorize_fastener)

    # Acara
    if 'Acara' in df.columns:
        df['Acara'] = df['Acara'].map(occasion_map).fillna('Other')
    # Rename non-English columns
    rename_dict = {
        'Dikirim Dari': 'shipped_from',
        'Tinggi Hak': 'heel_height',
        'Bahan': 'material',
        'Tipe Ujung Sepatu': 'toe_type',
        'Wide Fit': 'wide_fit',
        'Negara Asal': 'country_of_origin',
        'Model Sepatu': 'shoe_model',
        'Tipe Pengikat': 'fastener_type',
        'Acara': 'occasion'
    }

    df = df.rename(columns=rename_dict)
    return df


# ==================================================================
# Execution
# ==================================================================
if __name__ == "__main__":
    # Load data
    input_file = "/Users/claranatalies/Documents/year 4/FYP/COMP4501-FYP/cleaned_dataset_final/Sandal_updated_cleaned.csv"
    df_raw = pd.read_csv(input_file, low_memory=False)

    # Clean
    df_cleaned = clean_sandals(df_raw)

    # Print value counts for the shoe‑specific columns
    shoe_cols = ['Tinggi Hak', 'Bahan', 'Tipe Ujung Sepatu', 'Wide Fit',
                 'Negara Asal', 'Model Sepatu', 'Tipe Pengikat', 'Acara']
    for col in shoe_cols:
        if col in df_cleaned.columns:
            print(f"\n{col} after cleaning:")
            print(df_cleaned[col].value_counts())

    # Optionally save the cleaned data
    df_cleaned.to_csv('sandal_final_cleaned.csv', index=False)
    print("\nCleaning complete.")
```
