```python
import os
import pandas as pd

# Paths
BASE_FOLDER = r"C:\Users\Acer\Downloads\COMP4501-FYP"
INPUT_FOLDER = os.path.join(BASE_FOLDER, "cleaned_dataset_final_v2")
OUTPUT_FOLDER = os.path.join(BASE_FOLDER, "cleaned_dataset_revised_v3")

os.makedirs(OUTPUT_FOLDER, exist_ok=True)

# ---------- Processing ---------- #

for file in os.listdir(INPUT_FOLDER):
    if file.endswith(".csv"):
        file_path = os.path.join(INPUT_FOLDER, file)
        print(f"\nProcessing: {file}")

        try:
            df = pd.read_csv(file_path)

            # ---------- Case 1: Specific file ---------- #
            if file == "panci_final_cleaned.csv":
                if "review_count" in df.columns:
                    df = df.rename(columns={"review_count": "reviews"})
                    print("✔ Renamed review_count → reviews")

            # ---------- Case 2: Other files ---------- #
            else:
                cols_to_drop = ["review", "total_reviews"]
                df = df.drop(
                    columns=[col for col in cols_to_drop if col in df.columns],
                    errors="ignore"
                )
                print("✔ Dropped review / total_reviews if exists")

            # ---------- Show head ---------- #
            print(df.head())

            # ---------- Save ---------- #
            save_path = os.path.join(OUTPUT_FOLDER, file)
            df.to_csv(save_path, index=False)

            print(f"✅ Saved to: {save_path}")

        except Exception as e:
            print(f"❌ Error processing {file}: {e}")

print("\n🎯 All files processed successfully.")
```
