### Clean seller chat time

```python
import pandas as pd
import re

# -------------------------------------------------------------------
# 1. Clean seller_chat_time_reply to numeric hours
# -------------------------------------------------------------------

def clean_chat_time(val):
    if pd.isna(val):
        return None
    s = str(val).strip().lower()
    if s == 'hitungan jam':
        return 1.0
    if s == 'hitungan menit':
        return 0.5
    match = re.search(r'([\d,\.]+)\s*(jam|menit|hours?|hrs?)', s)
    if match:
        num_str = match.group(1).replace(',', '.')
        try:
            num = float(num_str)
        except:
            return None
        unit = match.group(2)
        if unit.startswith('jam') or unit.startswith('hour') or unit.startswith('hr'):
            return num
        elif unit.startswith('menit'):
            return num / 60.0
        else:
            return None
    try:
        return float(s)
    except:
        return None

df['seller_chat_time_reply'] = df['seller_chat_time_reply'].apply(clean_chat_time)

# -------------------------------------------------------------------
# 2. Clean seller_joined_date to numeric years (relative to 2026-03-21)
# -------------------------------------------------------------------

REF_DATE = pd.to_datetime('2026-03-21')

def parse_absolute(s):
    s_clean = re.sub(r'\s*GMT.*$', '', s)
    s_clean = re.sub(r'\s*\(.*$', '', s_clean)
    try:
        return pd.to_datetime(s_clean, errors='coerce')
    except:
        return pd.NaT

def relative_to_years(s):
    match = re.match(r'(\d+)\s+(tahun|bulan|hari)\s+lalu', s)
    if not match:
        return None
    num = int(match.group(1))
    unit = match.group(2)
    if unit == 'tahun':
        return float(num)
    elif unit == 'bulan':
        return num / 12.0
    else:  # hari
        return num / 365.25

def convert_seller_joined_date(val):
    if pd.isna(val):
        return None
    val_str = str(val).strip()
    if 'lalu' in val_str:
        years = relative_to_years(val_str)
        if years is not None:
            return years
    dt = parse_absolute(val_str)
    if pd.notna(dt):
        return (REF_DATE - dt).days / 365.25
    return None

df['seller_joined_date'] = df['seller_joined_date'].apply(convert_seller_joined_date)
```

### Clean Bahan (Materials)

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)

df['Bahan'].value_counts()
```

```python
import pandas as pd
import re

# ---------- 1. Normalization ---------- #
def normalize_text(x):
    if pd.isna(x):
        return ""
    
    x = str(x).lower()
    
    # unify separators to comma
    x = re.sub(r"[|/+]", ",", x)
    
    # remove non-alphanumeric except comma
    x = re.sub(r"[^a-z0-9, ]", " ", x)
    
    # clean spaces
    x = re.sub(r"\s+", " ", x).strip()
    
    return x


# ---------- 2. Standardization Map ---------- #
material_map = {
    "nilon": "nylon",
    "nylon": "nylon",
    "spandek": "spandex",
    "spandex": "spandex",
    "katun": "cotton",
    "cotton": "cotton",
    "poliester": "polyester",
    "polyester": "polyester",
    "kulit sintetis": "synthetic",
    "kulit": "leather",
    "kanvas": "canvas",
    "wol": "wool",
    "sintetis": "synthetic",
    "plastik": "plastic",
    "karet": "rubber",
    "mesh": "mesh",
}


# ---------- 3. Noise Words ---------- #
noise_words = [
    "good quality", "premium", "original", "high quality",
    "kualitas bagus", "luxury", "import"
]


# ---------- 4. Main Cleaning Function ---------- #
def clean_bahan(x):
    text = normalize_text(x)

    # remove noise
    for noise in noise_words:
        text = text.replace(noise, "")

    materials = text.split(",")

    cleaned = []

    for m in materials:
        m = m.strip()

        if not m:
            continue

        # map to standard material
        for key in material_map:
            if key in m:
                cleaned.append(material_map[key])
                break

    # remove duplicates
    cleaned = list(set(cleaned))

    # ---------- 5. Final Rule ---------- #
    if len(cleaned) == 0:
        return None
    elif len(cleaned) == 1:
        return cleaned[0]
    else:
        return "mix"

df["Bahan"] = df["Bahan"].apply(clean_bahan)
```

```python
import pandas as pd

def clean_negara_asal(x):
    if pd.isna(x):
        return "others"
    
    x = str(x).lower().strip()

    # normalize variations
    if "indonesia" in x or "lokal" in x:
        return "indonesia"
    
    elif "china" in x:
        return "china"
    
    elif "vietnam" in x:
        return "vietnam"
    
    elif "korea" in x:
        return "korea"
    
    else:
        return "others"

df["Negara Asal"] = df["Negara Asal"].apply(clean_negara_asal)
```

### Clean Panjang Kaos Kaki

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)

df['Panjang Kaos Kaki'].value_counts()
```

```python
import pandas as pd
import re

def clean_panjang_kaos_kaki(x):
    if pd.isna(x):
        return "others"
    
    x = str(x).lower()

    # normalize text
    x = re.sub(r"[^a-z0-9 ]", " ", x)
    x = re.sub(r"\s+", " ", x).strip()

    # ---------- RULES ---------- #

    # Mata kaki (ankle)
    if any(keyword in x for keyword in [
        "mata kaki", "ankle", "engkel", "low cut", "semata kaki"
    ]):
        return "ankle-length"

    # Setinggi betis (calf)
    elif any(keyword in x for keyword in [
        "betis", "3 4", "setengah betis"
    ]):
        return "calf-length"

    # Setinggi lutut (knee / above)
    elif any(keyword in x for keyword in [
        "lutut", "sepaha", "paha"
    ]):
        return "knee-length"

    # Others
    else:
        return "others"

df["Panjang Kaos Kaki"] = df["Panjang Kaos Kaki"].apply(clean_panjang_kaos_kaki)
```

### Jenis Kaos Kaki

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)

df['Jenis Kaos Kaki'].value_counts()
```

```python
import pandas as pd
import re

def clean_jenis_kaos_kaki(x):
    if pd.isna(x):
        return "others"
    
    x = str(x).lower()

    # normalize text
    x = re.sub(r"[^a-z0-9 ]", " ", x)
    x = re.sub(r"\s+", " ", x).strip()

    # ---------- RULES ---------- #

    # Olahraga (sports)
    if any(keyword in x for keyword in [
        "olahraga", "sport", "futsal", "bola", "skate", "oldschool"
    ]):
        return "sport"

    # Kantor / kerja / formal
    elif any(keyword in x for keyword in [
        "kantor", "kerja", "formal", "pdl", "pdh", "dinas", "tni", "polri", "work"
    ]):
        return "office"

    # Sekolah
    elif any(keyword in x for keyword in [
        "sekolah", "pramuka"
    ]):
        return "school"

    # Casual / everyday
    elif any(keyword in x for keyword in [
        "casual", "hangout", "fashion"
    ]):
        return "casual"

    # Others
    else:
        return "others"

df["Jenis Kaos Kaki"] = df["Jenis Kaos Kaki"].apply(clean_jenis_kaos_kaki)
```
